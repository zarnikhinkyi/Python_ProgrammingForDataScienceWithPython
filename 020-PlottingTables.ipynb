```python
%matplotlib inline
import matplotlib.pyplot as plt

data=[[1,2,3,4,5],[2,4,6,8,10],[1,3,5,7,9]]
cols=['ColOne','ColTwo','ColThree','ColFour','ColFive']
rows=['RowOne','RowTwo','RowThree']
colsc=['gray','gray','gray','gray','gray']
rowsc=['cyan','cyan','cyan']
colors=[['red','w','w','w','w'],['w','w','green','w','w'],['w','w','w','w','blue']]

fig,ax=plt.subplots(1,1)
ax.axis('off')
ax.set_aspect(0.1)
table=ax.table(cellText=data,colLabels=cols,rowLabels=rows,loc='upper center',cellLoc='center',colColours=colsc,rowColours=rowsc,cellColours=colors)
table.scale(1,20)

plt.savefig('pytable.png',format='png')
```
