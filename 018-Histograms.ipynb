```python
%matplotlib inline
import matplotlib.pyplot as plt

x1=[40,40,80,80,75,75,75,75,70,70,50,50,55,55,60,65,90,35,20]
x2=[20,20,30,30,45,45,50,50,65,60,65,75,75,80,80,90,90,35,85]

plt.hist(x1,5,label='x1', alpha=0.5,color='blue')
plt.hist(x2,10,label='x2', alpha=0.5,color='red')
plt.legend(loc='upper right')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x1=[40,40,80,80,75,75,75,75,70,70,50,50,55,55,60,65,90,35,20]
x2=[20,20,30,30,45,45,50,50,65,60,65,75,75,80,80,90,90,35,85]
num_bins=7
plt.hist(x1,num_bins,label='Maths', alpha=0.5,color='blue')
plt.hist(x2,num_bins,label='Physics', alpha=0.5,color='red')
plt.xlabel('Maths')
plt.ylabel('Physics')
plt.title('Student scores')
plt.legend(loc='upper right')
plt.show()
```
