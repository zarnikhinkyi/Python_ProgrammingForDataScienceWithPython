```python
import numpy as np
labels=['Samsung','Nokia','Apple','Oppo','Hiwei','Sony']
man=[10,5,4,2,10,10]
woman=[12,7,10,4,3,8]

fig,ax=plt.subplots()
width=0.4
x=np.arange(6)

ax.bar(x-width/2,man,width,label='Man')
ax.bar(x+width/2,woman,width,label='Women')
ax.legend()
ax.set_xticks(x)
ax.set_xticklabels(labels)
plt.show()
```
