```python
import numpy as np
rng=np.random.RandomState(0)

x=rng.randint(1,100,50)
y=rng.randint(1,100,50)

plt.xlabel('Maths')
plt.ylabel('Physics')
plt.title('Sudent score - Mtahs vs Physics')

plt.scatter(x,y,s=100,color='green',marker='v')
plt.show()
```

```python
import numpy as np
rng=np.random.RandomState(0)

x=rng.randint(1,100,50)
y=rng.randint(1,100,50)

sizes=10*np.add(x,y)

plt.xlabel('Maths')
plt.ylabel('Physics')
plt.title('Student score-Maths vs Physics')

plt.scatter(x,y,s=sizes, alpha=0.5)
plt.show()
```

```python
import numpy as np
rng=np.random.RandomState(0)

x=rng.randint(1,100,50)
y=rng.randint(1,100,50)

colors=rng.randint(1,100,50)
sizes=10*np.add(x,y)

plt.xlabel('Maths')
plt.ylabel('Physics')
plt.title('Student score-Maths vs Physics')

plt.scatter(x,y,c=colors,s=sizes, alpha=0.5)
plt.show()
```

```python
import numpy as np
rng=np.random.RandomState(0)

x=rng.randint(1,100,50)
y=rng.randint(1,100,50)

colors=rng.randint(1,100,50)
sizes=10*np.add(x,y)

plt.xlabel('Maths')
plt.ylabel('Physics')
plt.title('Student score-Maths vs Physics')

plt.scatter(x,y,c=colors,s=sizes, alpha=0.5, cmap='plasma')
plt.colorbar()
plt.show()
```
