```python
%matplotlib inline
import matplotlib.pyplot as plt
x=[1,3,5]
y=[1,3,5]
plt.plot(x,y)
plt.show()
```

```python
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Chart Example')
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,linestyle='dotted',linewidth=2.5,color='red')
plt.show()
```

```python
plt.plot(x,y,linestyle='dashed',linewidth=5,color='red')
plt.show()
```

```python
plt.plot(x,y,linestyle='dashdot',linewidth=5,color='red')
plt.show()
```

```python
x1=[1,3,5,7,9]
y1=[1,3,5,7,9]
x2=[10,8,6,4,2]
y2=[9,7,5,3,1]
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```python
x1=[1,3,5,7,9]
y1=[2,4,6,8,10]

plt.plot(x1,y1)
for a,b in zip(x1,y1):
    plt.text(a, b, str(a) + ',' + str(b))
plt.show()
```
