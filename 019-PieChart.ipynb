```python
%matplotlib inline
import matplotlib.pyplot as plt

data =[25,75]
plt.pie(data)
plt.axis('equal')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

data =[100,50,150]
plt.pie(data)
plt.axis('equal')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

data =[100,50,150]
lbl=['Cofee','Tea','Milk']
plt.pie(data,labels=lbl)
plt.title('Drink distribution')
plt.axis('equal')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

data =[100,50,150]
lbl=['Cofee','Tea','Milk']
col=['brown','orange','gray']
plt.pie(data,labels=lbl,colors=col)
plt.title('Drink distribution')
plt.axis('equal')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

data =[100,50,150]
lbl=['Cofee','Tea','Milk']
col=['brown','orange','gray']
spl=[0,0.1,0]
plt.pie(data,labels=lbl,colors=col,explode=spl,shadow=True)
plt.title('Drink distribution')
plt.axis('equal')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

data =[100,50,150]
lbl=['Cofee','Tea','Milk']
col=['brown','orange','gray']
spl=[0,0.1,0]
plt.pie(data,labels=lbl,colors=col,explode=spl,shadow=True,autopct='%1.2f%%')
plt.title('Drink distribution')
plt.axis('equal')
plt.show()
```
